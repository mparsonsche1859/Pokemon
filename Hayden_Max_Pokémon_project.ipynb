```python
#%%pycodestyle


import random as rand


class Pokemon:

    def __init__(self, name, type_, HP, moves, attack,
                 defense, speed, experience):
        self.name = name
        self.type_ = type_
        self.HP = HP
        self.attack = attack
        self.defense = defense
        self.speed = speed
        self.experience = experience
        self.moves = moves

    """This function calculates the damage of an attack and rut
    urns that damage"""
    def attack_(self, move, opponent):
        power = POWERS[move]
        for potentialmovetype, movelist in MOVES.items():
            if move in movelist:
                movetype = potentialmovetype
        typemultiplier = 1
        for opponenttype in POKEMONS[opponent]['Type']:
            if opponenttype in SUPER_EFFECTIVE[movetype]:
                typemultiplier *= 2
            if opponenttype in NOT_VERY_EFFECTIVE[movetype]:
                typemultiplier *= 0.5
        if self.speed > rand.randint(0, 511):
            print("A critical hit!")
            critmultiplier = 2
        else:
            critmultiplier = 1
        randmultiplier = rand.randint(85, 100)/100
        statmultiplier = self.attack/POKEMONS[opponent]['Defense']
        levelmultiplier = 2 * int(self.experience ** (1/3)) / 5 + 2
#         print('power', power)
#         print('typemultiplier', typemultiplier)
#         print('crit', critmultiplier)
#         print('rand', randmultiplier)
#         print('stat', statmultiplier)
#         print('lvl', levelmultiplier)
        print(f"-- BAM --")
        part = (levelmultiplier * power * statmultiplier / 50) * critmultiplier
        print(f"{move} does {part * randmultiplier * typemultiplier} damage!")
        return ((levelmultiplier * power * statmultiplier / 50) *
                critmultiplier * randmultiplier * typemultiplier)

    """This function allows a pokemon to take the given amount of damage"""
    def take_damage(self, damage):
        self.HP -= damage

    """This function calculates new experience and updates the level"""
    def update_level(self, opponent):
        if self.experience == 1:
            self.experience = 0
        newexperience = (self.experience + POKEMONS[opponent]['Exp'])
        print(f"Congratulations! your {self.name} gained ", end='')
        print(f"{POKEMONS[opponent]['Exp']} experience from the battle.")
        oldexp = int(round(self.experience ** (1/3), 3))
        newexp = int(round(newexperience**(1/3), 3))
        if oldexp != newexp:
            # print('EXP', self.experience)
            # print('NEWEXP', newexperience)
            # print('lvl', int(self.experience ** (1/3)))
            # print('newlvl', int(newexperience ** (1/3)))
            self.experience = newexperience

    def __str__(self):
        return self.name


class Pikachu(Pokemon):

    def __init__(self):
        super().__init__('Pikachu', ['Electric'], 35,
                         ['Thunder Shock',  'Double Kick', 'Thunderbolt'],
                         55, 40, 90, 1)


class Charizard(Pokemon):

    def __init__(self):
        super().__init__('Charizard', ['Fire', 'Flying'], 78,
                         ['Crunch', 'Ember', 'Scratch', 'Wing Attack'],
                         84, 78, 100, 1)


class Squirtle(Pokemon):

    def __init__(self):
        super().__init__('Squirtle', ['Water'], 44,
                         ['Tackle', 'Bubble', 'Bite'],
                         48, 65, 43, 1)


class Jigglypuff(Pokemon):

    def __init__(self):
        super().__init__('Jigglypuff', ['Normal', 'Fairy'], 115,
                         ['Pound', 'Body Slam', 'Double Slap'],
                         45, 20, 20, 1)


class Gengar(Pokemon):

    def __init__(self):
        super().__init__('Gengar', ['Ghost', 'Poison'], 60,
                         ['Lick', 'Smog', 'Dream Eater', 'Shadow Ball'],
                         65, 60, 110, 1)


class Magnemite(Pokemon):

    def __init__(self):
        super().__init__('Magnemite', ['Electric', 'Steel'], 25,
                         ['Tackle', 'Flash Cannon', 'Thunder Shock',
                          'Thunderbolt'], 35, 70, 45, 1)


class Bulbasaur(Pokemon):

    def __init__(self):
        super().__init__('Bulbasaur', ['Grass', 'Poison'], 45,
                         ['Tackle', 'Vine Whip', 'Razor Leaf'],
                         49, 49, 45, 1)


class Charmander(Pokemon):

    def __init__(self):
        super().__init__('Charmander', ['Fire'], 39,
                         ['Scratch', 'Ember', 'Fire Spin'],
                         52, 43, 65, 1)


class Beedrill(Pokemon):

    def __init__(self):
        super().__init__('Beedrill', ['Bug', 'Poison'], 65,
                         ['Peck', 'Twineedle', 'Rage',
                          'Fury Attack', 'Outrage'],
                         90, 40, 75, 1)


class Golem(Pokemon):

    def __init__(self):
        super().__init__('Golem', ['Rock', 'Ground'], 80,
                         ['Tackle', 'Rock Throw',
                          'Rock Slide', 'Earthquake'],
                         120, 130, 45, 1)


class Dewgong(Pokemon):

    def __init__(self):
        super().__init__('Dewgong', ['Water', 'Ice'], 90,
                         ['Aqua Jet', 'Ice Shard', 'Headbutt'],
                         70, 80, 70, 1)


class Hypno(Pokemon):

    def __init__(self):
        super().__init__('Hypno', ['Psychic'], 85,
                         ['Pound', 'Confusion', 'Dream Eater'],
                         73, 70, 67, 1)


class Cleffa(Pokemon):

    def __init__(self):
        super().__init__('Cleffa', ['Fairy'], 50,
                         ['Pound', 'Magical Leaf'],
                         25, 28, 15, 1)


class Cutiefly(Pokemon):

    def __init__(self):
        super().__init__('Cutiefly', ['Fairy', 'Bug'], 40,
                         ['Absorb', 'Fairy Wind', 'Struggle Bug',
                          'Draining Kiss'], 45, 40, 84, 1)
```
